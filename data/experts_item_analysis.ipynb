```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

```python
target_csv = pd.read_csv('/chronos_data/avirinchipur/reasoning_for_psych/kjell_prolific/depTextOnly.csv')
target_csv
```

```python
annotation_col_dict = {'Anhedonia': 'PHQ.SQ01.', 'Depressed mood': 'PHQ.SQ02.', 'Insomnia or Hypersomnia': 'PHQ.SQ03.', 'Fatigue': 'PHQ.SQ04.', 'Poor appetite or overeating': 'PHQ.SQ05.', 
                       'Worthlessness or Guilt': 'PHQ.SQ06.', 'Difficulty concentrating': 'PHQ.SQ07.', 'Psychomotor agitation or retardation': 'PHQ.SQ08.', 'Suicidal ideation': 'PHQ.SQ09.'}
```

## Veerle annotations

```python
veerle_annotation = pd.read_csv('./Item_level_scores_Veerle.csv')
veerle_annotation
```

```python
veerle_annotation[veerle_annotation[annotation_col_dict.keys()].isna().sum(1)>0]
```

```python
veerle_annotation = veerle_annotation.fillna(0)
```

```python
veerle_annotation[veerle_annotation.isna().sum(1)>0]
```

```python
veerle_annotation = veerle_annotation.rename(columns=annotation_col_dict)
veerle_annotation
```

```python
veerle_annotation['PHQ.SUM'] = veerle_annotation[[col for col in veerle_annotation.columns if 'PHQ.SQ' in col]].sum(1)
veerle_annotation
```

```python
x_y = pd.merge(target_csv[['user_id', 'phq_score']], veerle_annotation[['user_id', 'PHQ.SUM']], on='user_id')
veerle_scores, sr_scores = x_y['PHQ.SUM'].values, x_y['phq_score'].values
pearson_score, p_val = pearsonr(veerle_scores, sr_scores)
print ("Pearson Correlation between Veerle's scores and SR scores: {} (p={}; N={})".format(round(pearson_score, 3), round(p_val, 3), len(x_y)))
```

```python
# Run correlation between PHQ2 
x, y = veerle_annotation[['user_id', 'PHQ.SQ01.', 'PHQ.SQ02.']], target_csv[['user_id', 'PHQ.SQ01.', 'PHQ.SQ02.']]
x['PHQ2'] = x['PHQ.SQ01.'] + x['PHQ.SQ02.']
y['PHQ2'] = y['PHQ.SQ01.'] + y['PHQ.SQ02.']
x_y = pd.merge(x[['user_id', 'PHQ2']], y[['user_id', 'PHQ2']], on='user_id')
veerle_scores, sr_scores = x_y['PHQ2_x'].values, x_y['PHQ2_y'].values
pearson_score, p_val = pearsonr(veerle_scores, sr_scores)
print ("Pearson Correlation between Veerle's PHQ2 scores and SR scores: {} (p={}; N={})".format(round(pearson_score, 3), round(p_val, 3), len(x_y)))
```

```python
# Pearson correlation between the target csv scores and verrle scores for each of PHQ 9 items
item_item_corr = []
for col_name in [col for col in veerle_annotation.columns if 'PHQ.SQ' in col]:
    x_y = pd.merge(target_csv[['user_id', col_name]], veerle_annotation[['user_id', col_name]], on='user_id')
    sr_scores, veerle_scores = x_y['{}_x'.format(col_name)].values, x_y['{}_y'.format(col_name)].values
    pearson_score, p_val = pearsonr(veerle_scores, sr_scores)
    item_item_corr.append((col_name, round(pearson_score, 3), round(p_val, 3)))
    # print ("Pearson Correlation between Veerle's scores and SR scores for {}: {} (p={}; N={})".format(col, round(pearson_score, 3), round(p_val, 3), len(x_y)))
item_item_corr = pd.DataFrame(item_item_corr, columns=['item', 'pearson', 'p_val'])
item_item_corr
```
