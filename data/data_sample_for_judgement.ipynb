```python
## ipynb for sampling data for expert judgement
import pandas as pd
```

```python
ALL_DATA_FILE = '/chronos_data/avirinchipur/reasoning_for_psych/expts/responses/expt_gpt-4.dep_list_phq9items_score_classify_editted.csv'
CURRENT_DATA_SAMPLE_FILE = '/chronos_data/avirinchipur/reasoning_for_psych/kjell_prolific/depText_kjell_scoring.csv'
```

```python
all_data_df = pd.read_csv(ALL_DATA_FILE)
current_data_df = pd.read_csv(CURRENT_DATA_SAMPLE_FILE, encoding='latin1')
```

```python
print (all_data_df.shape)
```

```python
print (current_data_df.shape)
```

```python
current_data_df
```

```python
remaining_data_df = all_data_df[~all_data_df['user_id'].isin(current_data_df['user_id'])]
remaining_data_df.shape
```

```python
new_sample = remaining_data_df.sample(n=110, replace=False, random_state=42)[['user_id', 'user_text', 'target_value']]
new_sample
```

```python
new_sample['PHQ_tot_estimate'] = ''
new_sample
```

```python
new_sample = new_sample.rename(columns={'user_text':'dep_text'})
new_sample
```

```python
new_sample[['user_id', 'dep_text', 'PHQ_tot_estimate']].to_csv('./new_sample_for_experts.csv', index=False, encoding='utf-8')
```

```python
temp_df = pd.merge(all_data_df, current_data_df[['user_id', 'PHQ_tot_estimates_a', 'PHQ_tot_estimates_c']], on='user_id', how='inner')
for phq_estimate_col in ['PHQ_tot_estimates_a', 'PHQ_tot_estimates_c']:
    temp_current_df = temp_df[['user_id', 'user_text', phq_estimate_col]] 
    temp_current_df = temp_current_df.rename(columns={'user_text':'dep_text', phq_estimate_col:'PHQ_tot_estimate'})
    temp_combined_df = pd.concat([temp_current_df, new_sample], axis=0)
    temp_combined_df[['user_id', 'dep_text', 'PHQ_tot_estimate']].to_csv('./new_sample_for_experts_{}.csv'.format(phq_estimate_col), index=False, encoding='utf-8')
```
